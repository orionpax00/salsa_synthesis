```python
import os
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
data_man = np.load("data/all_asfamc/npsubjects/60/60_01.npy")
data_woman = np.load("data/all_asfamc/npsubjects/61/61_01.npy")
```

```python
fig = plt.figure()
ax = Axes3D(fig)
xm=[]
ym=[]
zm=[]
xw = []
yw = []
zw = []
for i in data_man[0]:
    xm.append(i[0])
    ym.append(i[1])
    zm.append(i[2])
    
for i in data_woman[0]:
    xw.append(i[0])
    yw.append(i[1])
    zw.append(i[2])
plt.plot(zw,xw,yw, "b.") ## plotting order should be Z,X,Y not X,Y,Z
plt.plot(zm,xm,ym, "r.") ## plotting order should be Z,X,Y not X,Y,Z
plt.show()
```
